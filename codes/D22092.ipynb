```python
import numpy as np
import matplotlib.pyplot as plt
import math
tol=1.0e-14
def my_exp(k,n=100,debug=False):
    y=0
    for i in range(0,n):
        temp=y
        y=y+k**i/np.math.factorial(i)
        if debug:
            if(abs(y-temp)<tol):
                print(f"seires is converged at i={i}")
                break
    return y
                
```

```python
x=2
my_exp(x,100,True)
```

```python
x=np.linspace(0,100,1000)
y=my_exp(x)
z=np.exp(x)
plt.plot(x,y,'-r',label="my_exp")
plt.plot(x,z,'-b',label="numpy_exp")
plt.legend()
plt.show()
```
